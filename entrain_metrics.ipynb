```python
# alt version, take means for each word cat
import numpy as np
score_dict = {}
for word_cat in mi_scores.dnm_scores.keys():
    score_list1 = []
    for dialog_id, spkrs in miami.dialog_spkrs.iteritems():
        score1 = mi_scores.dnm_scores[word_cat][spkrs[0]][0]  # score (diff only)
        score2 = mi_scores.dnm_scores[word_cat][spkrs[1]][0]  # score (diff only)
        score_list1.append(score1)
        score_list1.append(score2)
        
    word_cat_avg = np.mean(score_list1)
    score_dict[word_cat] = word_cat_avg
#     print '{}\t{}'.format(word_cat, word_cat_avg)
new_dict = sorted(score_dict, key=score_dict.get)
new_dict.reverse()
# print '\n*********\n'
for word_cat in new_dict:
    print '{}\t{}'.format(word_cat, score_dict[word_cat])
```
